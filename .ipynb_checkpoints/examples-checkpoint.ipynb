```python
import torch, gc
import torch.nn.functional as F
from transformers import CLIPTokenizer, CLIPTextModel
from diffusers import AutoencoderKL, DDPMScheduler, StableDiffusionPipeline, UNet2DConditionModel
from image_meta_dataset import ImageMetaDataset
from torchvision import transforms

# data preparation
img_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((256, 256), antialias = True),
    transforms.CenterCrop(256),
    transforms.RandomHorizontalFlip(),
    transforms.Normalize([0.5], [0.5])
])

comic_dataset = ImageMetaDataset(
    csv_file = '/scratch/project_2005600/political_comics/metadata.csv', 
    meta_columns = ['Title', 'tags_string'], 
    path_column = 'path',
    transform = img_transform
)

from torch.utils.data import DataLoader
dataloader = DataLoader(comic_dataset, batch_size = 32, shuffle = True)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# initilize diffusion models
tokenizer = CLIPTokenizer.from_pretrained(
    "CompVis/stable-diffusion-v1-4", 
    subfolder="tokenizer")

text_encoder = CLIPTextModel.from_pretrained(
    "CompVis/stable-diffusion-v1-4", 
    subfolder="text_encoder").to(device)

vae = AutoencoderKL.from_pretrained(
    "CompVis/stable-diffusion-v1-4", 
    subfolder="vae").to(device)

noise_scheduler = DDPMScheduler.from_pretrained(
    "CompVis/stable-diffusion-v1-4", 
    subfolder="scheduler")

text_encoder.requires_grad_(False)
vae.requires_grad_(False)

unet = UNet2DConditionModel.from_pretrained(
	pretrained_model_name_or_path = "CompVis/stable-diffusion-v1-4",
	subfolder="unet").to(device)

unet.train()

print('init')
```

```python
optimizer = torch.optim.AdamW(
    unet.parameters(),
    lr=1e-4)

n_steps = len(dataloader)
```

```python
epochs = 5
```

```python
for epoch in range(epochs):
    for i, data in enumerate(dataloader):
        images = data['image'].to(device)
        
        inputs = tokenizer(
            data['Title'], 
            max_length=tokenizer.model_max_length, 
            padding="max_length", 
            truncation=True, 
            return_tensors="pt")
        
        input_ids = inputs.input_ids.to(device)
        
        latents = vae.encode(images).latent_dist.sample()
        latents = latents * vae.config.scaling_factor
        noise = torch.randn_like(latents)
        batch_size = latents.shape[0]
        timesteps = torch.randint(0, 
                                  noise_scheduler.config.num_train_timesteps, 
                                  (batch_size,), 
                                  device=latents.device)
        
        timesteps = timesteps.long()
        noisy_latents = noise_scheduler.add_noise(latents, noise, timesteps)
        del latents
        torch.cuda.empty_cache()
        
        encoder_hidden_states = text_encoder(input_ids)[0]
        
        model_pred = unet(noisy_latents, timesteps, encoder_hidden_states).sample
        del noisy_latents, timesteps, encoder_hidden_states
        torch.cuda.empty_cache()

        # compute loss and gradients
        loss = F.mse_loss(model_pred.float(), noise.float(), reduction="mean")
        loss.backward()
        
        # update parameters
        optimizer.step()
        optimizer.zero_grad()
        
        # print status
        if i % 10 == 0:
            print(f'epoch: [{epoch+1}/{epochs}]')
            print(f'[{i+1}/{n_steps}]')    
            print(f'total loss: {loss.item()}')
        
        gc.collect()
        torch.cuda.empty_cache()
```

```python
# save fine-tuned diffusion model
pipeline = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    text_encoder=text_encoder,
    vae=vae,
    unet=unet
)

pipeline.save_pretrained(output_dir)
```
